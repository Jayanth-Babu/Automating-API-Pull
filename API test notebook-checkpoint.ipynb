```python
import pandas as pd
df = pd.read_csv(r'F:\PERSONAL\Data Analyst Projects\API\API.csv')
df
```

```python
# Now let's look at the coin trends over time

df2 = df.groupby('name', sort=False)[['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d']].mean()
df3 = df2.stack()
df4 = df3.reset_index()
df5 = df4.rename(columns = {'level_1':'perc_change', 0:'values'})
df5
```

```python
df5['perc_change'] = df5['perc_change'].replace(['quote.USD.percent_change_1h','quote.USD.percent_change_24h','quote.USD.percent_change_7d','quote.USD.percent_change_30d','quote.USD.percent_change_60d','quote.USD.percent_change_90d'],['1h','24h','7d','30d','60d','90d'])
df5
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.catplot(x='perc_change', y='values', hue='name', data=df5, kind='point')
plt.show()
```
